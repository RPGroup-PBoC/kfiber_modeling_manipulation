Studies
- Parameter sweeps:
    - s*: transition point (discrete choices: [0.5, 1, 1.5, 2, 2.5, 3])
    - Mk: moment at the kinetochore (wide range, to identify threshold gamma values)
    - gamma: thickness parameter
        - Look in the literature for plausible values
- Apply this idea to real data
    - Fixed s* at several different values
    - Optimize for gamma with a point moment at the kinetochore
    - Includes Rx in the model
- Use `n_div = 400` for final studies

- End-matching isn't good enough (maybe MCMC is the solution, or something simpler..)
    - Enforce no looping behavior
    - Take care of outlier cases
- Make representative shapes (place the points on the headmaps)
- Colored heatmaps with a threshold line
- Indicate regions of unacceptible behavior (the white regions currently)

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import scipy.optimize
from scipy.interpolate import interp1d
from tqdm.notebook import tqdm_notebook as tqdm

import numba
from numba.core import types
from numba.typed import Dict
import emcee

import sys
sys.path.append('../')
from functions import *
```

```python
float_array = types.float64[:]
float_val = types.float64

@numba.njit
def deflection_exact(param, gamma, Mk, s_star, L_contour, xF, Fext_y, n_div = 200):
    
    """
    gamma  : Thickness factor
    s_star : Contour distance from the kinetochore where
             k-fiber thinning happens
    """
    
    theta0, Fy = param
    
    ds = L_contour/n_div
    n_steps = n_div
    
    
    x_ls = np.zeros(n_steps, dtype=np.float64)
    y_ls = np.zeros(n_steps)
    M_ls = np.zeros(n_steps)
    theta_ls = np.zeros(n_steps)
    
    # Iteration index, after which thinning happens
    i_star = s_star/ds
    
    
    theta_ls[0] = theta0
    found_F = False
    yF = 0
    
    for i in range(n_steps-1):
         
        xF_sign = np.sign(xF) * np.sign(xF - x_ls[i])
        if xF_sign < 0 and found_F == False:
            found_F = True
            
            iF = i-1
            yF = interpolate_two_pt(xF, x_ls[iF], x_ls[iF+1], y_ls[iF], y_ls[iF+1])            
            
        M = Mk - Fy*x_ls[i]
        if found_F:
            M += -Fext_y*(x_ls[i]-xF)
        
        
        if i <= i_star:
            theta_mid = theta_ls[i] - (ds/2)*M/gamma
        else:
            theta_mid = theta_ls[i] - (ds/2)*M
            
        y_mid = y_ls[i] + ds/2*np.sin(theta_mid)
        x_mid = x_ls[i] + ds/2*np.cos(theta_mid)
        
        
        M_mid = Mk - Fy*x_mid
        if found_F:
            M_mid += -Fext_y*(x_mid-xF)
        
        if i <= i_star:
            theta_ls[i+1] = theta_ls[i] - ds*M_mid/gamma
        else:
            theta_ls[i+1] = theta_ls[i] - ds*M_mid
            
        x_ls[i+1] = x_ls[i] + ds*np.cos(theta_mid)
        y_ls[i+1] = y_ls[i] + ds*np.sin(theta_mid)
        M_ls[i+1] = M_mid
        
    # Calculate curvature
    kappa_ls = (theta_ls[:-1]-theta_ls[1:])/ds
    
    
    deflect_arrays = Dict.empty(
        key_type=types.unicode_type,
        value_type=float_array
    )
    
    deflect_numbers = Dict.empty(
        key_type=types.unicode_type,
        value_type=float_val
    )
    
    deflect_arrays['x_mod'] = x_ls.astype(np.float64)
    deflect_arrays['y_mod'] = y_ls.astype(np.float64)
    deflect_arrays['theta_mod'] = theta_ls.astype(np.float64)
    deflect_arrays['kappa_mod'] = kappa_ls.astype(np.float64)
    deflect_arrays['M_mod'] = M_ls.astype(np.float64)
    
    deflect_numbers['xF'] = np.float64(xF)
    deflect_numbers['yF'] = np.float64(yF)
    deflect_numbers['i_star'] = i_star
    
    return deflect_arrays, deflect_numbers
```

```python
@numba.njit
def error_fn(param, gamma, Mk, s_star, L_contour, xF, Fext_y):
    deflect_arrays, deflect_numbers = deflection_exact(param, gamma, Mk, s_star, L_contour, xF, Fext_y)

    M_mod = deflect_arrays["M_mod"]
    y_mod = deflect_arrays["y_mod"]
    
    M_right = M_mod[-1]
    y_right = y_mod[-1]
    
    error = abs(M_right)**2 + abs(y_right)**2
    
    return error
```

```python
def give_opt(gamma, Mk, s_star, L_contour, xF, Fext_y, n_steps=50, n_rand=100):
    
    gamma_init = 2
    Mk_init = -0.4
    s_star_init = 0.5
    L_contour_init = 10.0
    xF_init = 5.0
    Fext_y_init = 0.5
    
    param = [0.50875956, -0.20648158]
    
    gamma_seq = np.linspace(gamma_init, gamma, n_steps)
    Mk_seq = np.linspace(Mk_init, Mk, n_steps)
    s_star_seq = np.linspace(s_star_init, s_star, n_steps)
    L_contour_seq = np.linspace(L_contour_init, L_contour, n_steps)
    xF_seq = np.linspace(xF_init, xF, n_steps)
    Fext_y_seq = np.linspace(Fext_y_init, Fext_y, n_steps)
    
    
    for i in range(n_steps):
        np.random.seed(0)
        
        success = False
        for j in range(n_rand):
            gs = np.random.normal(loc=1, scale=0.0005, size=6)

            args = (gamma_seq[i]*gs[0], Mk_seq[i]*gs[1], s_star_seq[i]*gs[2], \
                    L_contour_seq[i]*gs[3], xF_seq[i]*gs[4], Fext_y_seq[i]*gs[5])
            
            output = scipy.optimize.minimize(error_fn, param, args, method="Nelder-Mead")
            
            if output.success == True:
                param = output.x
                success = True
                break
        
        if not success:
            break
        
    return param, success
```

```python
@numba.njit
def logpost_exact(param, gamma, Mk, s_star, L_contour, xF, Fext_y):
    
    error = error_fn(param, gamma, Mk, s_star, L_contour, xF, Fext_y)
    
    deflect_arrays, deflect_numbers = deflection_exact(param, gamma, Mk, s_star, L_contour, xF, Fext_y)

    theta_mod = deflect_arrays["theta_mod"]
    
    if np.sum(theta_mod < -0.75*np.pi) > 0:
        return -np.inf
    
    return -np.log(error)
```

```python
n_dim = 2
n_walkers = 5
n_steps = 4000

cols = ['theta0', 'Fy']



L_contour = 10
xF = 5
Fext_y = 0.5
# gamma = 1.5
gamma = 1.0
Mk = -1.0
s_star = 2.0

param_opt, success = give_opt(gamma, Mk, s_star, L_contour, xF, Fext_y)


# -------------- MCMC run --------------
p_init_set = []

for i in range(n_walkers):
    p_init_set.append(param_opt)

# eps = 1e-2*np.random.rand(n_walkers, n_dim)
shuffle = np.random.normal(loc=1, scale=0.0005, size=(n_walkers, n_dim))
# p_init_set = np.array(p_init_set) + eps
p_init_set = np.array(p_init_set)*shuffle

args = (gamma, Mk, s_star, L_contour, xF, Fext_y)
sampler = emcee.EnsembleSampler(n_walkers, n_dim, logpost_exact, args=args, threads=3)

_ = sampler.run_mcmc(p_init_set, n_steps, progress=True)
df_mcmc = sampler_to_dataframe(sampler, columns=cols)

p_opt_mcmc = df_mcmc[df_mcmc['lnprob']==df_mcmc['lnprob'].max()].values[0][:-2]
# --------------------------------------


# deflect_arrays, deflect_numbers = deflection_exact(param_opt, gamma, Mk, s_star, L_contour, xF, Fext_y)
deflect_arrays, deflect_numbers = deflection_exact(p_opt_mcmc, gamma, Mk, s_star, L_contour, xF, Fext_y)

x_mod = deflect_arrays["x_mod"]
y_mod = deflect_arrays["y_mod"]
s_mod, _ = contour_fn(x_mod, y_mod)

kappa_mod = deflect_arrays["kappa_mod"]



i_star = int(deflect_numbers['i_star'])
kappa_K = kappa_mod[0]
kappa_J = kappa_mod[1+i_star]


plt.plot(x_mod[s_mod <= s_star], y_mod[s_mod <= s_star], lw=2*gamma, color='k', solid_capstyle='butt')
plt.plot(x_mod, y_mod, lw=2, color='orange', solid_capstyle='butt')
plt.title('success: ' + str(success))
plt.axis('equal')
plt.show()
```

```python
theta0_thresh = np.radians(0)
```

```python
n_mesh = 10

gamma_vals = np.linspace(1.00, 5, n_mesh)
Mk_vals = np.linspace(0, -1.25, n_mesh)

gamma_mesh, Mk_mesh = np.meshgrid(gamma_vals, Mk_vals)

kappa_K_mat = np.zeros_like(gamma_mesh)
kappa_J_mat = np.zeros_like(gamma_mesh)
theta0_mat = np.zeros_like(gamma_mesh)


for j, Mk in tqdm(enumerate(Mk_vals)):
    for i, gamma in enumerate(gamma_vals):
        
        param_opt, success = give_opt(gamma, Mk, s_star, L_contour, xF, Fext_y)
        
        if success:
            # -------------------- Run MCMC --------------------
            p_init_set = []

            for _ in range(n_walkers):
                p_init_set.append(param_opt)

            randomness = np.random.normal(loc=1,scale=0.0005,size=(n_walkers, n_dim))
            p_init_set = np.array(p_init_set) * randomness

            args = (gamma, Mk, s_star, L_contour, xF, Fext_y)
            sampler = emcee.EnsembleSampler(n_walkers, n_dim, logpost_exact, args=args, threads=3)

            _ = sampler.run_mcmc(p_init_set, n_steps, progress=False)
            df_mcmc = sampler_to_dataframe(sampler, columns=cols)

            p_opt_mcmc = df_mcmc[df_mcmc['lnprob']==df_mcmc['lnprob'].max()].values[0][:-2]
            # ---------------------------------------------------


    #         deflect_arrays, deflect_numbers = \
    #             deflection_exact(param_opt, gamma, Mk, s_star, L_contour, xF, Fext_y)
            deflect_arrays, deflect_numbers = \
                deflection_exact(p_opt_mcmc, gamma, Mk, s_star, L_contour, xF, Fext_y)
        
        
            theta0_mod = deflect_arrays["theta_mod"][0]

            x_mod = deflect_arrays["x_mod"]
            y_mod = deflect_arrays["y_mod"]
            s_mod, _ = contour_fn(x_mod, y_mod)

            kappa_mod = deflect_arrays["kappa_mod"]

            i_star = int(deflect_numbers['i_star'])
            kappa_K = kappa_mod[0]
            kappa_J = kappa_mod[1+i_star]
        
        if success:
            kappa_K_mat[j,i] = kappa_K
            kappa_J_mat[j,i] = kappa_J
            theta0_mat[j,i] = theta0_mod
        else:
            kappa_K_mat[j,i] = None
            kappa_J_mat[j,i] = None
            
        plt.plot(x_mod, y_mod, color='k', alpha=0.3)
        
    plt.axis('equal')
    plt.title(Mk)
    plt.show()
```
