```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rotation(x_arr,y_arr,angle):
    """Rotates a point around the origin (0,0) in the clockwise direction. 
       Angle is in radians. Point is a tuple"""
    x_rot_arr, y_rot_arr = (x_arr*np.cos(angle)+y_arr*np.sin(angle),-x_arr*np.sin(angle)+y_arr*np.cos(angle))
    return x_rot_arr, y_rot_arr
```

```python
df = pd.read_pickle("noforce_Mkzero.pkl")
df.head()
```

```python
Rx_ls = df["Rx"].values
M0_ls = df["M0"].values
Ry_ls = df["Ry"].values

sns.swarmplot(data=[Rx_ls,M0_ls,Ry_ls], palette = ["red", "darkturquoise", "gold"], size = 9)
plt.errorbar([0,1,2],[np.mean(Rx_ls),np.mean(M0_ls),np.mean(Ry_ls)], \
             yerr = [np.std(Rx_ls)/np.sqrt(len(Rx_ls)),np.std(M0_ls)/np.sqrt(len(M0_ls)),np.std(Ry_ls)/np.sqrt(len(Ry_ls))], \
             elinewidth=6, fmt = '_',color = "k", marker = '_',markersize = 5, \
             capsize = 7, zorder = 10, capthick=1.5)
plt.xticks([0,1,2], ("Rx","M0","Ry"))
plt.axhline(y=0, color = "k", ls = "--")
# plt.ylim(-0.5,0.8)
plt.show()

print("Mean Rx = %f +/- %f"%(np.mean(Rx_ls)*1000,np.std(Rx_ls*1000)/np.sqrt(len(Rx_ls))))
print("Mean M0 = %f +/- %f"%(np.mean(M0_ls)*100,np.std(M0_ls*100)/np.sqrt(len(M0_ls))))
print("Mean Ry = %f +/- %f"%(np.mean(Ry_ls)*1000,np.std(Ry_ls*1000)/np.sqrt(len(Ry_ls))))
```

```python
angles = [0,10,20,30,40]
colors=["gold", "orange", "red", "maroon", "black"]
Rx_ls, Ry_ls, M0_ls, xlabels = [],[],[],[]
for i in range(len(angles)-1):
    angle_0 = np.radians(angles[i])
    angle_1 = np.radians(angles[i+1])
    df_angle = df[(df['theta0']>=angle_0) & (df['theta0']<angle_1)]
    if df_angle.empty:
        print("empty")
    else:
        Rx_ls.append(df_angle["Rx"].values*df_angle["L"].values**2)
        M0_ls.append(df_angle["M0"].values*df_angle["L"].values)
        Ry_ls.append(df_angle["Ry"].values*df_angle["L"].values**2)
        xlabels.append("%s-%s"%(angles[i],angles[i+1]))

plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
sns.swarmplot(data=Rx_ls, palette = ["red"]*len(Rx_ls), size = 9)
for i in range(len(Rx_ls)):
    plt.errorbar(i,np.mean(Rx_ls[i]), \
                 yerr = np.std(Rx_ls[i])/np.sqrt(len(Rx_ls[i])),\
                 elinewidth=6, fmt = '_',color = "k", marker = '_',markersize = 20, \
                 capsize = 7, zorder = 10, capthick=1.5)
plt.axhline(y=0, color = "k", ls = "--")
plt.xticks(list(range(len(Rx_ls))), (xlabels))
plt.title("Rx")

plt.subplot(1,3,2)
sns.swarmplot(data=M0_ls, palette = ["green"]*len(M0_ls), size = 9)
for i in range(len(Ry_ls)):
    plt.errorbar(i,np.mean(M0_ls[i]), \
                 yerr = np.std(M0_ls[i])/np.sqrt(len(M0_ls[i])),\
                 elinewidth=6, fmt = '_',color = "k", marker = '_',markersize = 20, \
                 capsize = 7, zorder = 10, capthick=1.5)
plt.axhline(y=0, color = "k", ls = "--")
plt.xticks(list(range(len(Rx_ls))), (xlabels))
plt.title("M0")

plt.subplot(1,3,3)
sns.swarmplot(data=Ry_ls, palette = ["blue"]*len(Ry_ls), size = 9)
for i in range(len(Ry_ls)):
    plt.errorbar(i,np.mean(Ry_ls[i]), \
                 yerr = np.std(Ry_ls[i])/np.sqrt(len(Ry_ls[i])),\
                 elinewidth=6, fmt = '_',color = "k", marker = '_',markersize = 20, \
                 capsize = 7, zorder = 10, capthick=1.5)
plt.axhline(y=0, color = "k", ls = "--")
plt.xticks(list(range(len(Rx_ls))), (xlabels))
plt.title("Ry")
plt.tight_layout()
plt.show()


nbins = 20
for i in range(len(angles)-1):
    x_mod_ls,y_mod_ls = [],[]
    
    angle_0 = np.radians(angles[i])
    angle_1 = np.radians(angles[i+1])
    df_angle = df[(df['theta0']>=angle_0) & (df['theta0']<angle_1)]

    if df_angle.empty:
        pass
    else:
        for index, row in df_angle.iterrows():
            x_mod = row["x_mod"]
            y_mod = row["y_mod"]
            L = x_mod[-1] - x_mod[0]
            x_mod=x_mod/L
            y_mod=y_mod/L
            x_mod_ls.append(x_mod)
            y_mod_ls.append(y_mod)
            plt.plot(x_mod,y_mod,color="grey",ls ='--',alpha = 0.5)
    x,y = [],[]
    [x.extend(el) for el in x_mod_ls] 
    [y.extend(el) for el in y_mod_ls] 
#     plt.scatter(x,y,color="k")
    x = np.array(x)
    y = np.array(y)
    
    n, _ = np.histogram(x, bins=nbins)
    sy, _ = np.histogram(x, bins=nbins, weights=y)
    sy2, _ = np.histogram(x, bins=nbins, weights=y*y)
    mean = sy / n
    std = np.sqrt(sy2/n - mean*mean)
    plt.errorbar((_[1:] + _[:-1])/2, mean, yerr=std, fmt = 'o-', color=colors[i])

plt.axhline(y=0,color="k")    
plt.axis("equal")
plt.show()
```

```python
angles = [0,10,15,25,35,50]
colors=["gold", "orange", "red", "maroon", "black"]
Rx_ls, Ry_ls, M0_ls, xlabels = [],[],[],[]
for i in range(len(angles)-1):
    df_angle = df[(df['PP-K angle (deg)']>=angles[i]) & (df['PP-K angle (deg)']<angles[i+1])]
    if df_angle.empty:
        print("empty")
    else:
        Rx,Ry = rotation(df_angle['Rx']*df_angle['L']**2,df_angle['Ry']*df_angle['L']**2,-np.radians(df_angle['PP-K angle (deg)']))
#         Rx,Ry = df_angle['Rx']*df_angle['L']**2,df_angle['Ry']*df_angle['L']**2
        Rx_ls.append(Rx)
        M0_ls.append(df_angle['Ry']*df_angle['L'])
        Ry_ls.append(Ry)
        xlabels.append("%s-%s"%(angles[i],angles[i+1]))

plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
sns.swarmplot(data=Rx_ls, palette = ["red"]*len(Rx_ls), size = 9)
for i in range(len(Rx_ls)):
    plt.errorbar(i,np.mean(Rx_ls[i]), \
                 yerr = np.std(Rx_ls[i])/np.sqrt(len(Rx_ls[i])),\
                 elinewidth=6, fmt = '_',color = "k", marker = '_',markersize = 20, \
                 capsize = 7, zorder = 10, capthick=1.5)
plt.axhline(y=0, color = "k", ls = "--")
plt.xticks(list(range(len(Rx_ls))), (xlabels))
plt.title("Rx")

plt.subplot(1,3,2)
sns.swarmplot(data=M0_ls, palette = ["green"]*len(M0_ls), size = 9)
for i in range(len(Ry_ls)):
    plt.errorbar(i,np.mean(M0_ls[i]), \
                 yerr = np.std(M0_ls[i])/np.sqrt(len(M0_ls[i])),\
                 elinewidth=6, fmt = '_',color = "k", marker = '_',markersize = 20, \
                 capsize = 7, zorder = 10, capthick=1.5)
plt.axhline(y=0, color = "k", ls = "--")
plt.xticks(list(range(len(Rx_ls))), (xlabels))
plt.title("M0")

plt.subplot(1,3,3)
sns.swarmplot(data=Ry_ls, palette = ["blue"]*len(Ry_ls), size = 9)
for i in range(len(Ry_ls)):
    plt.errorbar(i,np.mean(Ry_ls[i]), \
                 yerr = np.std(Ry_ls[i])/np.sqrt(len(Ry_ls[i])),\
                 elinewidth=6, fmt = '_',color = "k", marker = '_',markersize = 20, \
                 capsize = 7, zorder = 10, capthick=1.5)
plt.axhline(y=0, color = "k", ls = "--")
plt.xticks(list(range(len(Rx_ls))), (xlabels))
plt.title("Ry")
plt.tight_layout()
plt.show()


nbins = 20
for i in range(len(angles)-1):
    x_mod_ls,y_mod_ls = [],[]
    df_angle = df[(df['PP-K angle (deg)']>=angles[i]) & (df['PP-K angle (deg)']<angles[i+1])]
    if df_angle.empty:
        pass
    else:
        rot_angle = np.mean([angles[i+1],angles[i]])
        for index, row in df_angle.iterrows():
            x_mod = row["x_mod"]
            y_mod = row["y_mod"]
            L = x_mod[-1] - x_mod[0]
            x_mod_rot,y_mod_rot = rotation(x_mod,y_mod,np.radians(-rot_angle))
            x_mod_rot=x_mod_rot/L
            y_mod_rot=y_mod_rot/L
            x_mod_ls.append(x_mod_rot)
            y_mod_ls.append(y_mod_rot)
            plt.plot(x_mod_rot,y_mod_rot,color="grey",ls ='--',alpha = 0.5)
    x,y = [],[]
    [x.extend(el) for el in x_mod_ls] 
    [y.extend(el) for el in y_mod_ls] 
#     plt.scatter(x,y,color="k")
    x = np.array(x)
    y = np.array(y)
    
    n, _ = np.histogram(x, bins=nbins)
    sy, _ = np.histogram(x, bins=nbins, weights=y)
    sy2, _ = np.histogram(x, bins=nbins, weights=y*y)
    mean = sy / n
    std = np.sqrt(sy2/n - mean*mean)
    plt.errorbar((_[1:] + _[:-1])/2, mean, yerr=std, fmt = 'o-', color=colors[i])

plt.axhline(y=0,color="k")    
plt.axis("equal")
plt.show()
```
