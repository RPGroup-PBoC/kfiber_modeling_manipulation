```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import scipy.optimize
from scipy.interpolate import interp1d
from tqdm.notebook import tqdm_notebook as tqdm

import numba
from numba.core import types
from numba.typed import Dict
import emcee

import sys
sys.path.append('../')
from functions import *
```

```python
float_array = types.float64[:]
float_val = types.float64

@numba.njit
def deflection_exact(param, gamma, Mk, s_star, L_contour, xF, Fext_y, n_div = 400):
    
    """
    gamma     : Thickness factor.
    Mk        : Moment at the kinetochore.
    s_star    : Contour distance from the kinetochore where
                k-fiber thinning happens.
    L_contour : Contour length of the k-fiber.
    xF        : x-position of force application.
    Fext_y    : External force pointing vertically up.
    """
    
    theta0, Fy = param
    
    ds = L_contour/n_div
    n_steps = n_div
    
    x_ls = np.zeros(n_steps, dtype=np.float64)
    y_ls = np.zeros(n_steps)
    M_ls = np.zeros(n_steps)
    theta_ls = np.zeros(n_steps)
    
    # Iteration index, after which thinning happens
    i_star = s_star/ds
    
    
    theta_ls[0] = theta0
    found_F = False
    yF = 0
    
    for i in range(n_steps-1):
         
        xF_sign = np.sign(xF) * np.sign(xF - x_ls[i])
        if xF_sign < 0 and found_F == False:
            found_F = True
            
            iF = i-1
            yF = interpolate_two_pt(xF, x_ls[iF], x_ls[iF+1], y_ls[iF], y_ls[iF+1])            
            
        M = Mk - Fy*x_ls[i]
        if found_F:
            M += -Fext_y*(x_ls[i]-xF)
        
        
        if i <= i_star:
            theta_mid = theta_ls[i] - (ds/2)*M/gamma
        else:
            theta_mid = theta_ls[i] - (ds/2)*M
            
        y_mid = y_ls[i] + ds/2*np.sin(theta_mid)
        x_mid = x_ls[i] + ds/2*np.cos(theta_mid)
        
        
        M_mid = Mk - Fy*x_mid
        if found_F:
            M_mid += -Fext_y*(x_mid-xF)
        
        if i <= i_star:
            theta_ls[i+1] = theta_ls[i] - ds*M_mid/gamma
        else:
            theta_ls[i+1] = theta_ls[i] - ds*M_mid
            
        x_ls[i+1] = x_ls[i] + ds*np.cos(theta_mid)
        y_ls[i+1] = y_ls[i] + ds*np.sin(theta_mid)
        M_ls[i+1] = M_mid
        
    # Calculate curvature
    kappa_ls = -(theta_ls[1:]-theta_ls[:-1])/ds
    
    
    deflect_arrays = Dict.empty(
        key_type=types.unicode_type,
        value_type=float_array
    )
    
    deflect_numbers = Dict.empty(
        key_type=types.unicode_type,
        value_type=float_val
    )
    
    deflect_arrays['x_mod'] = x_ls.astype(np.float64)
    deflect_arrays['y_mod'] = y_ls.astype(np.float64)
    deflect_arrays['theta_mod'] = theta_ls.astype(np.float64)
    deflect_arrays['kappa_mod'] = kappa_ls.astype(np.float64)
    deflect_arrays['M_mod'] = M_ls.astype(np.float64)
    
    deflect_numbers['xF'] = np.float64(xF)
    deflect_numbers['yF'] = np.float64(yF)
    deflect_numbers['i_star'] = i_star
    
    return deflect_arrays, deflect_numbers
```

```python
@numba.njit
def error_fn(param, gamma, Mk, s_star, L_contour, xF, Fext_y):
    deflect_arrays, deflect_numbers = deflection_exact(param, gamma, Mk, s_star, L_contour, xF, Fext_y)

    M_mod = deflect_arrays["M_mod"]
    y_mod = deflect_arrays["y_mod"]
    
    M_right = M_mod[-1]
    y_right = y_mod[-1]
    
    error = abs(M_right)**2 + abs(y_right)**2
    
    return error
```

```python
def give_opt(gamma, Mk, s_star, L_contour, xF, Fext_y, n_steps=50, n_rand=100):
    """
    Obtains the optimal set of [theta0, Fy] pair correponding to the given
    set of parameters. Since convergence is highly sensitive to the initial
    guess, this is achieved by sequentially updating the parameters until they
    match the input values and using the converged pair at one step as the
    initial pair in the next step.
    
    The output `success` is assigned the value `False` if optimization fails
    during the iterative search.
    """
    
    gamma_init = 2
    Mk_init = -0.4
    s_star_init = 0.5
    L_contour_init = 10.0
    xF_init = 5.0
    Fext_y_init = 0.5
    
    param = [0.50875956, -0.20648158]
    
    gamma_seq = np.linspace(gamma_init, gamma, n_steps)
    Mk_seq = np.linspace(Mk_init, Mk, n_steps)
    s_star_seq = np.linspace(s_star_init, s_star, n_steps)
    L_contour_seq = np.linspace(L_contour_init, L_contour, n_steps)
    xF_seq = np.linspace(xF_init, xF, n_steps)
    Fext_y_seq = np.linspace(Fext_y_init, Fext_y, n_steps)
    
    
    for i in range(n_steps):
        np.random.seed(0)
        
        success = False
        for j in range(n_rand):
            gs = np.random.normal(loc=1, scale=0.0005, size=6)

            args = (gamma_seq[i]*gs[0], Mk_seq[i]*gs[1], s_star_seq[i]*gs[2], \
                    L_contour_seq[i]*gs[3], xF_seq[i]*gs[4], Fext_y_seq[i]*gs[5])
            
            output = scipy.optimize.minimize(error_fn, param, args, method="Nelder-Mead")
            
            if output.success == True:
                param = output.x
                success = True
                break
        
        if not success:
            break
        
    return param, success
```

```python
@numba.njit
def logpost_exact(param, gamma, Mk, s_star, L_contour, xF, Fext_y):
    
    error = error_fn(param, gamma, Mk, s_star, L_contour, xF, Fext_y)
    
    deflect_arrays, deflect_numbers = deflection_exact(param, gamma, Mk, s_star, L_contour, xF, Fext_y)

    x_mod = deflect_arrays["x_mod"]
    y_mod = deflect_arrays["y_mod"]
    theta_mod = deflect_arrays["theta_mod"]
    
    if np.sum(theta_mod < -0.75*np.pi) > 0:
        return -np.inf
    
    if np.sum(theta_mod > np.pi/2) > 0:
        return -np.inf
    
    if np.sum(x_mod[0:-1] > x_mod[1:]) > 0:
        return -np.inf

    
    return -2*np.log(error)
```

```python
n_dim = 2 # Number of parameters
n_walkers = 5 # Number of MCMC walkers
n_steps = 200 # Number of MCMC steps
cols = ['theta0', 'Fy'] # Names of the two parameters

L_contour = 10 # K-fiber contour length
xF = 5 # Position of force application (x-distance from kinetochore)
Fext_y = 0.5 # External force pointing up
gamma = 3.0 # Thickness parameter
Mk = 0.0 # Moment at the kinetochore
s_star = 2.5 # Thickening length along the contour
```

```python
# -------------- "Gradient descent" --------------
param_opt, _ = give_opt(gamma, Mk, s_star, L_contour, xF, Fext_y)
```

```python
# -------------- MCMC run --------------
np.random.seed(0)
n_try_max = 2e6
args = (gamma, Mk, s_star, L_contour, xF, Fext_y)

p_init_set = []

success = False

if logpost_exact(param_opt, *args) != -np.inf:
    for i in range(n_walkers):
        p_init_set.append(param_opt)
    success = True
    
else:
    n_try = 0
    while True:
        param_init = np.random.normal(0, 2, size=2)

        logpost = logpost_exact(param_init, *args)
        if logpost != -np.inf:
            p_init_set.append(param_init)
        n_try += 1
        
        if n_try >= n_try_max:
            success = False
            break
        
        if len(p_init_set) == n_walkers:
            success = True
            break

if success:
    shuffle = np.random.normal(loc=1, scale=0.0005, size=(n_walkers, n_dim))
    p_init_set = np.array(p_init_set)*shuffle

    sampler = emcee.EnsembleSampler(n_walkers, n_dim, logpost_exact, args=args, threads=3)

    _ = sampler.run_mcmc(p_init_set, n_steps, progress=True)
    df_mcmc = sampler_to_dataframe(sampler, columns=cols)

    p_opt_mcmc = df_mcmc[df_mcmc['lnprob']==df_mcmc['lnprob'].max()].values[0][:-2]
```

```python
if success:
    # deflect_arrays, deflect_numbers = deflection_exact(param_opt, gamma, Mk, s_star, L_contour, xF, Fext_y)
    deflect_arrays, deflect_numbers = deflection_exact(p_opt_mcmc, gamma, Mk, s_star, L_contour, xF, Fext_y)

    x_mod = deflect_arrays["x_mod"]
    y_mod = deflect_arrays["y_mod"]
    s_mod, _ = contour_fn(x_mod, y_mod)
    
    L = x_mod[-1]

    kappa_mod = deflect_arrays["kappa_mod"]
    
    i_star = int(deflect_numbers['i_star'])
    kappa_K = kappa_mod[0]
    kappa_J = kappa_mod[1+i_star]

    plt.plot(L-x_mod[s_mod <= s_star], y_mod[s_mod <= s_star], lw=3*gamma, color='orange', solid_capstyle='butt')
    plt.plot(L-x_mod, y_mod, lw=3, color='orange', solid_capstyle='butt')
    plt.xticks(fontsize=11)
    plt.yticks(fontsize=11)
    plt.xlabel('x (µm)', fontsize=12, font='Lucida Grande')
    plt.ylabel('y (µm)', fontsize=12, font='Lucida Grande')
    plt.axis('equal')
    plt.savefig('fig_outputs/variable_EI_gamma%.2f_Mk%.2f_sstar%.2f.pdf'%(gamma, np.abs(Mk), s_star))
    plt.show()
    
    
    plt.plot(L-x_mod[:-1], kappa_mod, color='k')
    plt.axhline(0, color='k', lw=0.5, ls='--')
    plt.xlabel('x (µm)', fontsize=14, font='Lucida Grande')
    plt.ylabel(r'curvature (µm$^{-1}$)', fontsize=14, font='Lucida Grande')
    plt.xticks(fontsize=11)
    plt.yticks(fontsize=11)
    plt.savefig('fig_outputs/variable_EI_curvature_gamma%.2f_Mk%.2f_sstar%.2f.pdf'%(gamma, np.abs(Mk), s_star))
    plt.show()
```

### Parametric studies

```python
theta0_thresh = np.radians(0)
```

```python
n_mesh = 50
n_try_max = 2e6

gamma_vals = np.linspace(1.00, 5, n_mesh)
Mk_vals = np.linspace(0, -1.4, n_mesh)

gamma_mesh, Mk_mesh = np.meshgrid(gamma_vals, Mk_vals)

kappa_K_mat = np.full(gamma_mesh.shape, None)
kappa_J_mat = np.full(gamma_mesh.shape, None)
theta0_mat = np.full(gamma_mesh.shape, None)

for j, Mk in tqdm(enumerate(Mk_vals)):
    for i, gamma in enumerate(gamma_vals):

        # -------------- "Gradient descent" --------------
        param_opt, _ = give_opt(gamma, Mk, s_star, L_contour, xF, Fext_y)
        

        # -------------- MCMC run --------------
        np.random.seed(0)
        
        args = (gamma, Mk, s_star, L_contour, xF, Fext_y)

        p_init_set = []

        success = False

        if logpost_exact(param_opt, *args) != -np.inf:
            for _ in range(n_walkers):
                p_init_set.append(param_opt)
            success = True

        else:
            n_try = 0
            while True:
                param_init = np.random.normal(0, 2, size=2)

                logpost = logpost_exact(param_init, *args)
                if logpost != -np.inf:
                    p_init_set.append(param_init)
                n_try += 1

                if n_try >= n_try_max:
                    success = False
                    break

                if len(p_init_set) == n_walkers:
                    success = True
                    break

        if success:
            shuffle = np.random.normal(loc=1, scale=0.0005, size=(n_walkers, n_dim))
            p_init_set = np.array(p_init_set)*shuffle

            sampler = emcee.EnsembleSampler(n_walkers, n_dim, logpost_exact, args=args, threads=3)

            _ = sampler.run_mcmc(p_init_set, n_steps, progress=False)
            df_mcmc = sampler_to_dataframe(sampler, columns=cols)

            p_opt_mcmc = df_mcmc[df_mcmc['lnprob']==df_mcmc['lnprob'].max()].values[0][:-2]

            
            
            deflect_arrays, deflect_numbers = deflection_exact(p_opt_mcmc, gamma, Mk, \
                                                               s_star, L_contour, xF, Fext_y)

            x_mod = deflect_arrays["x_mod"]
            y_mod = deflect_arrays["y_mod"]
            s_mod, _ = contour_fn(x_mod, y_mod)

            theta0_mod = deflect_arrays["theta_mod"][0]
            kappa_mod = deflect_arrays["kappa_mod"]

            i_star = int(deflect_numbers['i_star'])
            kappa_K = kappa_mod[0]
            kappa_J = kappa_mod[1+i_star]

            
#             print(kappa_K, kappa_J)
            
            if np.abs(y_mod[-1])/L_contour < 1e-2:
                
                kappa_K_mat[j,i] = kappa_K
                kappa_J_mat[j,i] = kappa_J
                theta0_mat[j,i] = theta0_mod

#                 plt.plot(x_mod, y_mod, color='k', alpha=0.3)
            
#     plt.title('Mk: %.2f'%(Mk,))
#     plt.xlabel('x (um)')
#     plt.ylabel('x (um)')
#     plt.axis('equal')
#     plt.show()
```

```python
mask_exclude_theta0 = np.full(gamma_mesh.shape, False)
mask_exclude_kappaJ = np.full(gamma_mesh.shape, False)

for i in range(n_mesh):
    for j in range(n_mesh):
        if theta0_mat[i,j]:
            if theta0_mat[i,j] < theta0_thresh:
                mask_exclude_theta0[i,j] = True
        if kappa_J_mat[i,j]:
            if kappa_J_mat[i,j] > 0:
                mask_exclude_kappaJ[i,j] = True
```

```python
kappa_K_mat[mask_exclude_theta0] = None
kappa_K_mat[mask_exclude_kappaJ] = None
kappa_J_mat[mask_exclude_theta0] = None
kappa_J_mat[mask_exclude_kappaJ] = None
```

```python
kappa_ratio_mat = np.full(gamma_mesh.shape, None)
for i in range(n_mesh):
    for j in range(n_mesh):
        if kappa_J_mat[i,j] and kappa_K_mat[i,j]:
            kappa_ratio_mat[i,j] = kappa_J_mat[i,j]/kappa_K_mat[i,j]
```

```python
vmin = 1 - np.abs(1-kappa_ratio_mat[kappa_ratio_mat != None]).max()
vmax = 1 + np.abs(1-kappa_ratio_mat[kappa_ratio_mat != None]).max()

plt.contourf(gamma_mesh, Mk_mesh, kappa_ratio_mat, levels=50, cmap=plt.cm.RdYlBu, vmin=vmin, vmax=vmax)
cbar = plt.colorbar()
cbar.ax.tick_params(labelsize=11) 
plt.contour(gamma_mesh, Mk_mesh, kappa_ratio_mat, levels=[1.00], zorder=1, colors='k', linewidths=2.0)

# plt.xlabel(r'thickness parameter, $\gamma$', fontsize=12, font='Lucida Grande')
# plt.ylabel('moment at the kinetochore,\n' r'$M_k\,/\,EI$ $\left(\mathrm{µm}^{-2}\right)$', 
#            fontsize=12, font='Lucida Grande')

# plt.xlim([1, 3.5])
plt.ylim([-1.5, 0.0])

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

### Summary plot

```python
n_mesh = 30
n_try_max = 2e6

Mk_vals = np.linspace(0, -1.5, n_mesh)

def find_critical(s_star, gamma_min, gamma_max):
    
    gamma_mid = 0.5*(gamma_min + gamma_max)
    
    kappa_ratios = []
    
    for j, Mk in enumerate(Mk_vals):
        
        # -------------- "Gradient descent" --------------
        param_opt, _ = give_opt(gamma_mid, Mk, s_star, L_contour, xF, Fext_y)

        
        # -------------- MCMC run --------------
        np.random.seed(0)

        args = (gamma_mid, Mk, s_star, L_contour, xF, Fext_y)

        p_init_set = []

        success = False

        if logpost_exact(param_opt, *args) != -np.inf:
            for _ in range(n_walkers):
                p_init_set.append(param_opt)
            success = True

        else:
            n_try = 0
            while True:
                param_init = np.random.normal(0, 2, size=2)

                logpost = logpost_exact(param_init, *args)
                if logpost != -np.inf:
                    p_init_set.append(param_init)
                n_try += 1

                if n_try >= n_try_max:
                    success = False
                    break

                if len(p_init_set) == n_walkers:
                    success = True
                    break

        if success:
            shuffle = np.random.normal(loc=1, scale=0.0005, size=(n_walkers, n_dim))
            p_init_set = np.array(p_init_set)*shuffle

            sampler = emcee.EnsembleSampler(n_walkers, n_dim, logpost_exact, args=args, threads=3)

            _ = sampler.run_mcmc(p_init_set, n_steps, progress=False)
            df_mcmc = sampler_to_dataframe(sampler, columns=cols)

            p_opt_mcmc = df_mcmc[df_mcmc['lnprob']==df_mcmc['lnprob'].max()].values[0][:-2]



            deflect_arrays, deflect_numbers = deflection_exact(p_opt_mcmc, gamma_mid, Mk, \
                                                               s_star, L_contour, xF, Fext_y)

            x_mod = deflect_arrays["x_mod"]
            y_mod = deflect_arrays["y_mod"]
            s_mod, _ = contour_fn(x_mod, y_mod)

            theta0_mod = deflect_arrays["theta_mod"][0]
            kappa_mod = deflect_arrays["kappa_mod"]

            i_star = int(deflect_numbers['i_star'])
            kappa_K = kappa_mod[0]
            kappa_J = kappa_mod[1+i_star]
            
            if np.abs(y_mod[-1])/L_contour < 1e-2:
                
                if theta0_mod >= theta0_thresh:
                    if kappa_J < 0:
                        kappa_ratios.append(kappa_J/kappa_K)
    
    if np.abs(kappa_ratios[-1] - 1) < 1e-2:
        return gamma_mid
    
    if kappa_ratios[-1] > 1.0:
        return find_critical(s_star, gamma_min, gamma_mid)
    else:
        return find_critical(s_star, gamma_mid, gamma_max)
```

```python
gamma_min = 1.0
gamma_max = 5.0

s_star_ls = np.linspace(0.5, 2.5, 11)
gamma_crit_ls = []

for s_star in tqdm(s_star_ls):
    
    gamma_crit = find_critical(s_star, gamma_min, gamma_max)
    gamma_crit_ls.append(gamma_crit)
```

```python
plt.figure(figsize=(5, 3.5))
plt.scatter(s_star_ls, gamma_crit_ls)
plt.plot(s_star_ls, gamma_crit_ls)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```
