```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.optimize
import scipy.signal
from scipy import interpolate
import glob
import re
from tqdm import tqdm_notebook as tqdm 
import matplotlib.image as mpimg
import seaborn as sns
```

```python
def rotation(x_arr,y_arr,angle):
    """Rotates a point around the origin (0,0) in the clockwise direction. 
       Angle is in radians. Point is a tuple"""
    x_rot_arr, y_rot_arr = (x_arr*np.cos(angle)+y_arr*np.sin(angle),-x_arr*np.sin(angle)+y_arr*np.cos(angle))
    return x_rot_arr, y_rot_arr


def contour_fn(x,y):    
    s = np.zeros(len(x))
    for i in range(1,len(s)):
        dx = x[i] - x[i-1]
        dy = y[i] - y[i-1]
        s[i] = s[i-1] + np.sqrt(dx**2+dy**2)
    L_contour = s[-1]
    return s, L_contour 
```

```python
def deflection(x,param,L):
    Rx, Ry = param
    M0=Ry*L
    if Rx == 0:
        y = L/2*(M0 - Ry*L/3)*x - M0/2*x**2 + Ry/6*x**3
    elif Rx < 0.0:
        k = np.sqrt(-Rx)
        y = Ry/Rx*(x-L*np.sinh(k*x)/np.sinh(k*L)) - M0/Rx*((1-np.exp(-k*x)) - (1-np.exp(-k*L))*np.sinh(k*x)/np.sinh(k*L))
    else:
        k = np.sqrt(Rx)
        y = M0/Rx*np.cos(k*x) + (M0/Rx*(1-np.cos(k*L)) - Ry/Rx*L)/np.sin(k*L)*np.sin(k*x) + Ry/Rx*x - M0/Rx
#     print(M0)
    return y

def res_fn(param,x_dat,y_dat,L):
    y_mod = deflection(x_dat,param,L)
    return y_dat - y_mod
```

```python
def deflection_exact(param, x_dat_ls, y_dat_ls, rot_angle, n_div=100, length_conserved=True, \
                     method='midpoint', Rx_zero=False, Ry_zero=False,force_PP = False):
    
    Rx, Ry, theta0 = param
    
    if Rx_zero:
        Rx = 0
    if Ry_zero:
        Ry = 0
        
    L = x_dat_ls[-1] - x_dat_ls[0] 
    M0 = Ry*L
    
    if force_PP:
        Ry = -np.tan(rot_angle)*Rx
        
    if length_conserved:
        s_dat_ls,L_contour = contour_fn(x_dat_ls,y_dat_ls)        
        n_steps = n_div+2
        ds = L_contour/n_div
        s_ls = ds*np.arange(n_steps)
    
    else:
        L = x_dat_ls[-1] - x_dat_ls[0] 
        n_steps = n_div*4
        ds = L/n_div
        found = False
    
    theta_ls = np.zeros(n_steps)
    theta_ls[0] = theta0
    x_ls = np.zeros(n_steps)
    y_ls = np.zeros(n_steps)

    if method=='midpoint':
        
        for i in range(1,n_steps):            
            
            M = M0 + Rx*y_ls[i-1] - Ry*x_ls[i-1]
            
            theta_mid = theta_ls[i-1] - ds/2*M
            y_mid = y_ls[i-1] + ds/2*np.sin(theta_mid)
            x_mid = x_ls[i-1] + ds/2*np.cos(theta_mid)
            
            M_mid = M0 + Rx*y_mid - Ry*x_mid
            
            theta_ls[i] = theta_ls[i-1] - ds*M_mid
            x_ls[i] = x_ls[i-1] + ds*np.cos(theta_mid)
            y_ls[i] = y_ls[i-1] + ds*np.sin(theta_mid)
            
            if not length_conserved:
                if x_ls[i] >= L:
                    x_ls = x_ls[0:i+1]
                    y_ls = y_ls[0:i+1]
                    theta_ls = theta_ls[0:i+1]

                    found = True
                    break  
                
    elif method=='euler':
        for i in range(1,n_steps):
            x_ls[i] = x_ls[i-1] + ds*np.cos(theta_ls[i-1])
            y_ls[i] = y_ls[i-1] + ds*np.sin(theta_ls[i-1])

            M = M0 + Rx*y_ls[i-1] - Ry*x_ls[i-1]
            
            dtheta = - ds*M
            theta_ls[i] = theta_ls[i-1] + dtheta
            
            if not length_conserved:
                if x_ls[i] >= L:
                    x_ls = x_ls[0:i+1]
                    y_ls = y_ls[0:i+1]
                    theta_ls = theta_ls[0:i+1]

                    found = True
                    break  

    else:
        raise ValueError('Incorrect method specified.')
    
    if length_conserved:
        fx = interpolate.interp1d(s_ls, x_ls) 
        fy = interpolate.interp1d(s_ls, y_ls) 
        x_output = fx(s_dat_ls)
        y_output = fy(s_dat_ls)
    
    else:
        x_output = x_dat_ls
        f = interpolate.interp1d(x_ls, y_ls,fill_value="extrapolate")
        y_output = f(x_output)
#     print(M0)
    return x_output, y_output, x_ls, y_ls 



def res_fn_exact(param, x_dat_ls, y_dat_ls, rot_angle, length_conserved=True, end_match=False, force_PP = False):

    x_mod,y_mod,_,_ = deflection_exact(param, x_dat_ls, y_dat_ls, rot_angle, length_conserved=length_conserved, \
                                       force_PP=force_PP)
    res = np.sqrt((x_mod-x_dat_ls)**2 + (y_mod-y_dat_ls)**2)

    if end_match:
        res*=(1+5000*abs(y_mod[-1])/y_dat_ls.max())**2

    return res


```

```python
path = "/Users/poojasuresh/Google Drive/Modeling the spindle mechanics/_paper/dat/WT_unmanip/"
cells = sorted(glob.glob("%s/*_before.tif"%path))

for cell in cells:
    spindle = cell.split("/")[-1].split(".")[0]
    image = mpimg.imread(cell)
    profiles = glob.glob("%s/%s_*"%(path,spindle))
    
    ind_poles = [profiles.index(i) for i in profiles if 'poles' in i]
    profiles.pop(ind_poles[0])
    pole_csv = glob.glob("%s/%s_poles.csv"%(path,spindle))[0]

    for profile in profiles:
        kfiber_number = profile.split("/")[-1].split(".")[0].split("_")[-1]
        print(spindle,kfiber_number)
        fig, ax = plt.subplots(1,3,figsize = (14, 4))

        df_profile = pd.read_csv(profile)
        df_pole = pd.read_csv(pole_csv)
        
        x_dat_ls = np.array(df_profile["X"].values)
        y_dat_ls = np.array(df_profile["Y"].values)
        y_dat_ls = scipy.signal.savgol_filter(y_dat_ls, 21, 3) # NEED TO STUDY THE CORRECT WINDOW SIZE, currently 51

        ax[0].imshow(image)
        ax[0].plot(x_dat_ls,y_dat_ls, color = 'k', label = 'exp')

        x_mean = np.mean(x_dat_ls)
        y_mean = np.mean(y_dat_ls)

        P1x = np.array(df_pole["X"].values)[0]
        P1y = np.array(df_pole["Y"].values)[0]
        P2x = np.array(df_pole["X"].values)[1]
        P2y = np.array(df_pole["Y"].values)[1]
        
        PP_vector = [P2x-P1x, P2y-P1y]
        K_vector = [x_dat_ls[-1]-x_dat_ls[0], y_dat_ls[-1]-y_dat_ls[0]]
        unit_PP_vector = PP_vector / np.linalg.norm(PP_vector)
        unit_K_vector = K_vector / np.linalg.norm(K_vector)
        dot_product = np.dot(unit_PP_vector, unit_K_vector)
        PP_K_angle = np.arccos(dot_product)
        if PP_K_angle > np.pi/2:
            PP_K_angle = np.pi - PP_K_angle
        
        dist_P1 = np.sqrt((x_mean - P1x)**2+(y_mean - P1y)**2)
        dist_P2 = np.sqrt((x_mean - P2x)**2+(y_mean - P2y)**2)

        if dist_P2 < dist_P1:
            pole = "pole_right"
            x_dat_ls = -x_dat_ls
        else:
            pole = "pole_left"
                        
        angle = np.arctan((y_dat_ls[-1] - y_dat_ls[0]) / (x_dat_ls[-1] - x_dat_ls[0]))
        x_dat_ls,y_dat_ls = rotation(x_dat_ls,y_dat_ls,angle)
        
        # Translation to start at origin
        x_dat_ls -= x_dat_ls[0]
        y_dat_ls -= y_dat_ls[0]

        # Flipping to ensure a positive initial slope
        theta0 = np.arctan2(y_dat_ls[3]-y_dat_ls[0],x_dat_ls[3]-x_dat_ls[0])
        if theta0 < 0.0:
            y_dat_ls = -y_dat_ls
            angle = -angle

        # Convert from pixel to micron units based on scope camera calibration
        is_2017 = re.search('2017',profile)
        if is_2017:
            scope_calibration = 1/0.0657
        else:
            scope_calibration = 1/0.105

        x_dat_ls = x_dat_ls/scope_calibration
        y_dat_ls = y_dat_ls/scope_calibration

        # P-K distance
        L = x_dat_ls[-1] - x_dat_ls[0]
#         print("P-K distance (L) =  %.2f um" % L)

        # Initial angle
        theta_init = np.arctan2(y_dat_ls[3]-y_dat_ls[0],x_dat_ls[3]-x_dat_ls[0])
#         print("initial angle =  %.2f degrees" % np.degrees(theta_init))

        # Contour length
        s_dat_ls = np.zeros(len(x_dat_ls))
        for i in range(1,len(s_dat_ls)):
            dx = x_dat_ls[i] - x_dat_ls[i-1]
            dy = y_dat_ls[i] - y_dat_ls[i-1]
            s_dat_ls[i] = s_dat_ls[i-1] + np.sqrt(dx**2+dy**2)

        L_contour = s_dat_ls[-1]
#         print("k-fiber length = %.2f um" % L_contour)

        force_PP = False
        error_min = 1e10
        for Ry_init in tqdm(np.logspace(-2,0.5,10)/L):
            param_init = np.array([0.0, Ry_init,theta_init])
            args = (x_dat_ls,y_dat_ls,angle, True, False, force_PP)

            output = scipy.optimize.least_squares(res_fn_exact, param_init, args=args, method="lm")
            p_opt_exact = output.x
            num_eval = output.nfev

            if error_min > output.cost:
                error_min = output.cost
                Ry_init_opt = Ry_init

        param_init = np.array([0.01, Ry_init_opt,theta_init])
        args = (x_dat_ls, y_dat_ls, angle, True, False, force_PP)

        output = scipy.optimize.least_squares(res_fn_exact, param_init, args=args, method="lm")
        p_opt_exact = output.x
        num_eval = output.nfev

        param_init = p_opt_exact
        args = (x_dat_ls, y_dat_ls, angle, False, True, force_PP)

        output = scipy.optimize.least_squares(res_fn_exact, param_init, args=args, method="lm")
        p_opt_exact = output.x

#         print("number of evaluations = %s"% num_eval)

        Rx_opt_exact = p_opt_exact[0]
        Ry_opt_exact = p_opt_exact[1]
        theta0_opt_exact = p_opt_exact[2]
#         print(p_opt_exact)
        
        param_init = np.array([0.0, Ry_init_opt])
        param_extra = L
        args = (x_dat_ls,y_dat_ls,param_extra)

        output = scipy.optimize.least_squares(res_fn, param_init, args=args, method="lm")
        p_opt = output.x
        
        Rx_opt = p_opt[0]
#         M0_opt = p_opt[1]
        Ry_opt = p_opt[1]
        
        x_mod_ls = deflection_exact(p_opt_exact,x_dat_ls,y_dat_ls,angle,force_PP=force_PP)[0]
        y_mod_ls = deflection_exact(p_opt_exact,x_dat_ls,y_dat_ls,angle,force_PP=force_PP)[1]
        
        ax[1].plot(x_dat_ls,y_dat_ls, '.', color = 'k', lw = 2, label = 'exp')
        ax[1].plot(x_mod_ls,y_mod_ls, '--', color = 'blue', label = 'model (numerical)')
        ax[1].plot(x_dat_ls,deflection(x_dat_ls,p_opt,L), color = 'orange', \
                   label = 'model (small-angle)', lw = 8, alpha = 0.4)
        ax[1].legend()
        ax[1].set_xlabel('x (um)')
        ax[1].set_ylabel('y (um)')
        ax[1].axis('equal')        
        
        table_data = [[round(i,5) for i in p_opt[0:2]],[round(i,5) for i in p_opt_exact[0:2]]]
        table = ax[2].table(cellText=table_data, colLabels=['Rx', 'Ry'], loc='center', 
                          rowLabels=['small-angle', 'numerical'], cellLoc='center')
        table.scale(0.7,4)
        table.auto_set_font_size(False)
        table.set_fontsize(12)
        ax[2].axis('off')
        
        res_exact = res_fn_exact(p_opt_exact,x_dat_ls,y_dat_ls,L)
        err_exact = np.sqrt(np.mean(res_exact**2))
#         print("large angle", Rx_opt_exact,M0_opt_exact,Ry_opt_exact,err_exact)
        
        res = res_fn(p_opt,x_dat_ls,y_dat_ls,L)
        err = np.sqrt(np.mean(res**2))
#         print("small angle", Rx_opt,M0_opt,Ry_opt, err)
        
        y_diff_max = 0
        for q in range(len(x_dat_ls)):
            y_diff = abs(y_dat_ls[q] - y_mod_ls[q])
            if y_diff > y_diff_max:
                y_diff_max = y_diff
                
#         x_dat_ls = x_dat_ls.tolist()
#         y_dat_ls = y_dat_ls.tolist()
#         x_mod_ls = x_mod_ls.tolist()
#         y_mod_ls = y_mod_ls.tolist()
    
        df_fin = df_fin.append({'Spindle': spindle,
                        'Kfiber': kfiber_number,
                        'Pole': pole, 
                        'PP-K angle (deg)': np.degrees(PP_K_angle), 
                        'x_dat':x_dat_ls,
                        'y_dat':y_dat_ls,
                        'x_mod': x_mod_ls,
                        'y_mod': y_mod_ls,
                        'L': L,
                        'Rx':Rx_opt_exact,
                        'M0': Ry_opt_exact*L,
                        'Ry': Ry_opt_exact,
                        'theta0':theta0_opt_exact,
                        'largest-y-diff (error)':y_diff_max, 
                        'Root-mean-square(error)':err_exact},ignore_index=True)

        plt.tight_layout()
        plt.show()
```
