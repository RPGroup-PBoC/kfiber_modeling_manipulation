```python
import numpy as np
from matplotlib.pyplot import cm
import matplotlib.pyplot as plt
import pandas as pd
import scipy.optimize
from scipy import interpolate
import scipy.signal
import glob
import re
```

## Define functions

```python
def rotation(x_arr,y_arr,angle):
    """Rotates a point around the origin (0,0) in the clockwise direction. 
       Angle is in radians. Point is a tuple"""
    x_rot_arr, y_rot_arr = (x_arr*np.cos(angle)+y_arr*np.sin(angle),-x_arr*np.sin(angle)+y_arr*np.cos(angle))
    return x_rot_arr, y_rot_arr


def deflection_exact_contour(s_output,param,L_contour,xF,n_div=100): ## change xF to sF
    
    Rx, M0, Ry, theta0, F, xF_var = param
    F = np.abs(F)
    
    if np.abs(xF_var - xF) < 0.5: # based on the needle diameter from the data being within 1 um
        xF = xF_var
    
    n_steps = n_div+2
    ds = L_contour/n_div
    theta_ls = np.zeros(n_steps)
    theta_ls[0] = theta0
    
    x_ls = np.zeros(n_steps)
    y_ls = np.zeros(n_steps)
    s_ls = ds*np.arange(n_steps)
    M_ls = np.zeros(n_steps)
    
    found = False
    found_F = False
    yF = 0
    thetaF = np.pi/2
    iF = 0
    for i in range(1,n_steps):

        if x_ls[i-1] <= xF:
            M = M0 + Rx*y_ls[i-1] - Ry*x_ls[i-1]
            iF = i-1
        else:
            if not found_F:
                f = interpolate.interp1d([x_ls[iF],x_ls[iF+1]], [y_ls[iF],y_ls[iF+1]])
                yF = f(xF)
                found_F = True

                thetaF = np.pi/2 + np.arctan2(y_ls[iF+1]-y_ls[iF], x_ls[iF+1]-x_ls[iF])
                Fx = F * np.cos(thetaF)
                Fy = F * np.sin(thetaF)

            M = M0 + (Rx+Fx)*y_ls[i-1] - (Ry+Fy)*x_ls[i-1] \
                - Fx*yF + Fy*xF

        theta_mid = theta_ls[i-1] - ds/2*M
        y_mid = y_ls[i-1] + ds/2*np.sin(theta_mid)
        x_mid = x_ls[i-1] + ds/2*np.cos(theta_mid)

        if x_mid <= xF:
            M_mid = M0 + Rx*y_mid - Ry*x_mid
            iF = i-1
        else:
            if not found_F:
                f = interpolate.interp1d([x_ls[iF],x_mid], [y_ls[iF],y_mid])
                yF = f(xF)                     
                found_F = True

                thetaF = np.pi/2 + np.arctan2(y_mid-y_ls[iF], x_mid-x_ls[iF])
                Fx = F * np.cos(thetaF)
                Fy = F * np.sin(thetaF)
            M_mid = M0 + (Rx+Fx)*y_mid - (Ry+Fy)*x_mid \
                - Fx*yF + Fy*xF            

        theta_ls[i] = theta_ls[i-1] - ds*M_mid
        x_ls[i] = x_ls[i-1] + ds*np.cos(theta_mid)
        y_ls[i] = y_ls[i-1] + ds*np.sin(theta_mid)
        M_ls[i] = M_mid 
    
#     print(s_ls[-1], 'end')
    fx = interpolate.interp1d(s_ls, x_ls) 
    fy = interpolate.interp1d(s_ls, y_ls) 
    x_output = fx(s_output)
    y_output = fy(s_output)
    return x_output,y_output,yF, M_ls


def res_fn_exact_contour(param,s_dat,x_dat,y_dat,L_contour,xF):
    x_mod,y_mod,yF,M_ls = deflection_exact_contour(s_dat,param,L_contour,xF)
    res = np.sqrt((x_mod-x_dat)**2 + (y_mod-y_dat)**2)
    return res

def deflection_exact_xy(x_output,param,L,xF,n_div=100): 
    
    Rx, M0, Ry, theta0, F, xF_var = param
    F = np.abs(F)
    
    if np.abs(xF_var - xF) < 0.5:
        xF = xF_var
    
    n_steps = n_div*4
    ds = L/n_div
    theta_ls = np.zeros(n_steps)
    theta_ls[0] = theta0
    
    x_ls = np.zeros(n_steps)
    y_ls = np.zeros(n_steps)
    M_ls = np.zeros(n_steps)
    
    found = False
    found_F = False
    yF = 0
    thetaF = np.pi/2
    iF = 0

    for i in range(1,n_steps):

        if x_ls[i-1] <= xF:
            M = M0 + Rx*y_ls[i-1] - Ry*x_ls[i-1]
            iF = i-1
        else:
            if not found_F:
                f = interpolate.interp1d([x_ls[iF],x_ls[iF+1]], [y_ls[iF],y_ls[iF+1]])
                yF = f(xF)
                found_F = True

                thetaF = np.pi/2 + np.arctan2(y_ls[iF+1]-y_ls[iF], x_ls[iF+1]-x_ls[iF])
                Fx = F * np.cos(thetaF)
                Fy = F * np.sin(thetaF)

            M = M0 + (Rx+Fx)*y_ls[i-1] - (Ry+Fy)*x_ls[i-1] \
                - Fx*yF + Fy*xF

        theta_mid = theta_ls[i-1] - ds/2*M
        y_mid = y_ls[i-1] + ds/2*np.sin(theta_mid)
        x_mid = x_ls[i-1] + ds/2*np.cos(theta_mid)
#             print(x_ls[i-1],x_mid)

        if x_mid <= xF:
            M_mid = M0 + Rx*y_mid - Ry*x_mid
            iF = i-1
        else:
            if not found_F:
                f = interpolate.interp1d([x_ls[iF],x_mid], [y_ls[iF],y_mid])
                yF = f(xF)                     
                found_F = True

                thetaF = np.pi/2 + np.arctan2(y_mid-y_ls[iF], x_mid-x_ls[iF])
                Fx = F * np.cos(thetaF)
                Fy = F * np.sin(thetaF)
            M_mid = M0 + (Rx+Fx)*y_mid - (Ry+Fy)*x_mid \
                - Fx*yF + Fy*xF            

        theta_ls[i] = theta_ls[i-1] - ds*M_mid
        x_ls[i] = x_ls[i-1] + ds*np.cos(theta_mid)
        y_ls[i] = y_ls[i-1] + ds*np.sin(theta_mid)

        M_ls[i] = M_mid

        if x_ls[i] >= L:
            x_ls = x_ls[0:i+1]
            y_ls = y_ls[0:i+1]
            theta_ls = theta_ls[0:i+1]
            M_ls = M_ls[0:i+1]

            found = True
            break 

#     for i in range(1,len(x_ls)):
#         if x_ls[i] < x_ls[i-1]:
#             x_ls[i] = x_ls[i-1]
            
#     if x_ls[-1] < L:
#         y_ls = y_ls*L/x_ls[-1]
#         x_ls = x_ls*L/x_ls[-1]*(1+1e-8)
        
    f = interpolate.interp1d(x_ls, y_ls,fill_value="extrapolate")
    y_output = f(x_output)
    return y_output,yF,thetaF,x_ls,y_ls,M_ls


def res_fn_xy(param,x_dat,y_dat,L,xF):
    y_mod = deflection_exact_xy(x_dat,param,L,xF)[0]
    res = (y_dat - y_mod)
    return res

def res_fn_endmatch(param,x_dat,y_dat,L,xF):
    y_mod = deflection_exact_xy(x_dat,param,L,xF)[0]
    res = (y_dat - y_mod)*(1+1000*abs(y_mod[-1])/y_dat.max())**2 #*np.exp(100*abs(y_mod[-1])/y_dat.max())
    return res
```

## Importing data (only the k-fiber in the final timepoint of the manipulation)

```python
cells = sorted(glob.glob("../../dat/WT_manip/20*"))

fin_track_ls = []
for cell in cells:
    kfibers = sorted(glob.glob(cell+"/kfiber_*.csv"))
    final_kfiber = kfibers[-1]
    fin_track_ls.append(final_kfiber)
```

## Storing model outputs

```python
df_fin = pd.DataFrame(columns = ["Spindle", "x_dat", "y_dat", "x_mod", "y_mod", "M_mod", "L", \
                                 "Rx", "M0", "Ry", "Fx", "Fy", "xF", "yF", "theta0", \
                                 "largest-y-diff (error)", "Quality of fit (error)"])
```

```python
for n,kfib in enumerate(fin_track_ls):
    cell = cells[n]
    spindle = cell.split("/")[-1].split(".")[0]
    df_dat = pd.read_csv(kfib,engine="python")
    
    x_dat_ls = df_dat["X"].values
    y_dat_ls = df_dat["Y"].values

    angle = np.arctan((y_dat_ls[-1] - y_dat_ls[0]) / (x_dat_ls[-1] - x_dat_ls[0]))
    x_dat_ls,y_dat_ls = rotation(x_dat_ls,y_dat_ls,angle)

    # Translation to start at origin
    x_dat_ls -= x_dat_ls[0]
    y_dat_ls -= y_dat_ls[0]

    # Flipping to ensure a positive initial slope
    theta0 = np.arctan2(y_dat_ls[3]-y_dat_ls[0],x_dat_ls[3]-x_dat_ls[0])
    if theta0 < 0.0:
        y_dat_ls = -y_dat_ls
        angle = -angle
#         print("Angle of P-K rotation =  %.2f degrees" %np.degrees(angle))

    # Convert from pixel to micron units based on scope camera calibration
    is_2017 = re.search('2017',cell)
    if is_2017:
        scope_calibration = 1/0.0657
    else:
        scope_calibration = 1/0.105

    x_dat_ls = x_dat_ls/scope_calibration
    y_dat_ls = y_dat_ls/scope_calibration

    # P-K distance
    L = x_dat_ls[-1] - x_dat_ls[0]
#         print("P-K distance (L) = %.2f um" % L)

    # Initial angle
    theta_init = np.arctan2(y_dat_ls[3]-y_dat_ls[0],x_dat_ls[3]-x_dat_ls[0])
#         print("initial angle =  %.2f degrees" % np.degrees(theta_init))

    y_needle = np.max(y_dat_ls)
    x_needle = x_dat_ls[np.where(y_dat_ls == np.max(y_dat_ls))]
    xF = x_needle[0]

    # Contour length
    s_dat_ls = np.zeros(len(x_dat_ls))
    for i in range(1,len(s_dat_ls)):
        dx = x_dat_ls[i] - x_dat_ls[i-1]
        dy = y_dat_ls[i] - y_dat_ls[i-1]
        s_dat_ls[i] = s_dat_ls[i-1] + np.sqrt(dx**2+dy**2)

    L_contour = s_dat_ls[-1]
#         print("k-fiber length = %.2f um" % L_contour)

    error_min = 1e10
    for M0_init in np.logspace(-2,1,13)/L:
#         param_init = np.array([0.0, 0.0, Ry_init, theta_init,0.0,0.0])
        param_init = np.array([0.0, M0_init, 0.0, theta_init,0.0, xF])
        args = (s_dat_ls,x_dat_ls,y_dat_ls,L_contour,xF)

        output = scipy.optimize.least_squares(res_fn_exact_contour, param_init, args=args, method="lm")
        p_opt = output.x
        num_eval = output.nfev

        yF = deflection_exact_contour(s_dat_ls,p_opt,L_contour,xF)[-2]

        if error_min > output.cost:
            error_min = output.cost
            M0_init_opt = M0_init

    param_init = np.array([0.0, M0_init_opt, 0.0, theta_init,0.0, xF])
    args = (s_dat_ls,x_dat_ls,y_dat_ls,L_contour,xF)

    output = scipy.optimize.least_squares(res_fn_exact_contour, param_init, args=args, method="lm")
    p_opt = output.x

    yF = deflection_exact_contour(s_dat_ls,p_opt,L_contour,xF)[-2]


    param_init = p_opt
    args = (x_dat_ls,y_dat_ls,L,xF)

    output = scipy.optimize.least_squares(res_fn_xy, param_init, args=args, method="lm")
    p_opt_xy = output.x

    param_init = p_opt_xy
    args = (x_dat_ls,y_dat_ls,L,xF)

    output = scipy.optimize.least_squares(res_fn_endmatch, param_init, args=args, method="lm")
    p_opt_xy = output.x

    deflection_exact_xy_output = deflection_exact_xy(x_dat_ls,p_opt_xy,L,xF)

    _,yF,thetaF,x_mod_ls,y_mod_ls,M_mod_ls = deflection_exact_xy_output

    plt.plot(x_dat_ls,y_dat_ls, color="orange", lw = 8)
    plt.plot(x_mod_ls,y_mod_ls,color = "blue", lw = 3)
    plt.xticks(fontsize= 15)
    plt.yticks(fontsize= 15)

    Rx_opt, M0_opt, Ry_opt, theta0_opt, F_opt, xF_opt = p_opt_xy
    Fx_opt = F_opt * np.cos(thetaF)
    Fy_opt = F_opt * np.sin(thetaF)

    lbl=''
    plt.axvline(xF, color='k', ls='--', alpha=0.4)
    if np.abs(xF_opt-xF) < 0.5:
        xF = xF_opt
        lbl = 'optimized xF'

    quiver_scale=1
#     plt.quiver(0,0,Rx_opt,Ry_opt, color='k',scale=quiver_scale, zorder = 20)
    plt.quiver(xF,yF,Fx_opt,abs(Fy_opt), color='k',scale=quiver_scale, zorder = 20, label=lbl)
#     plt.quiver(L,0,-(Rx_opt+Fx_opt),-(Ry_opt+Fy_opt), color='k',scale=quiver_scale, zorder = 20)
#     plt.xlabel('x (um)')
#     plt.ylabel('y (um)')
    plt.axis('equal')
#     plt.legend()
    
    y_diff_max = 0
    for q in range(len(x_dat_ls)):
        y_diff = abs(y_dat_ls[q] - y_mod_ls[q])
        if y_diff > y_diff_max:
            y_diff_max = y_diff

    res_exact = res_fn_xy(p_opt_xy,x_dat_ls,y_dat_ls,L,xF)
    err_exact = np.sqrt(np.mean(res_exact**2))
    
    df_fin = df_fin.append({'Spindle': spindle,
                    'x_dat':x_dat_ls,
                    'y_dat':y_dat_ls,
                    'x_mod': x_mod_ls,
                    'y_mod': y_mod_ls,
                    'M_mod': M_mod_ls,
                    'L': L,
                    'Rx':Rx_opt,
                    'M0': M0_opt,
                    'Ry': Ry_opt,
                    'Fx': Fx_opt,
                    'Fy': abs(Fy_opt),
                    'xF': xF,
                    'yF': yF,                            
                    'theta0':theta0_opt,
                    'largest-y-diff (error)':y_diff_max, 
                    'Root-mean-square(error)':err_exact},ignore_index=True)

    plt.show()
```
