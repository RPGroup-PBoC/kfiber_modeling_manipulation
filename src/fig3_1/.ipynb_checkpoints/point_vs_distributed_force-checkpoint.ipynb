```python
def deflection_exact_distributed_F(param, rhoF, ell_needle, sF, L_contour, n_div=100):
    Ry, theta0 = param
    
    ds = L_contour/n_div
    n_steps = n_div+2
    
    x_ls = np.zeros(n_steps)
    y_ls = np.zeros(n_steps)
    theta_ls = np.zeros(n_steps)
    M_ls = np.zeros(n_steps)
    
    theta_ls[0] = theta0

    Rx = 0
    M0 = 0
    
    dF = rhoF*ds
    n_segments = int(round(ell_needle/ds))
    
    segment_counter = 0
    
    thetaF_ls = []
    xF_ls = []
    yF_ls = []
    Fx_ls = []
    Fy_ls = []
    
    i_start = int((sF-0.5*ell_needle)/ds)
    
    for i in range(n_steps-1):
        
        if i > i_start and segment_counter < n_segments:        
            xF = x_ls[i-1]
            yF = y_ls[i-1]
            thetaF = np.pi/2 + np.arctan2(y_ls[i]-y_ls[i-1], x_ls[i]-x_ls[i-1])

            Fx = dF * np.cos(thetaF)
            Fy = dF * np.sin(thetaF)

            xF_ls.append(xF)
            yF_ls.append(yF)
            thetaF_ls.append(thetaF)
            Fx_ls.append(Fx)
            Fy_ls.append(Fy)

            segment_counter += 1
            
        M = M0 + Rx*y_ls[i] - Ry*x_ls[i]
        
        for k in range(segment_counter):
            Fx = Fx_ls[k]
            Fy = Fy_ls[k]
            xF = xF_ls[k]
            yF = yF_ls[k]
            M += Fx*(y_ls[i]-yF) - Fy*(x_ls[i]-xF)
        
        
        theta_mid = theta_ls[i] - ds/2*M
        y_mid = y_ls[i] + ds/2*np.sin(theta_mid)
        x_mid = x_ls[i] + ds/2*np.cos(theta_mid)
        
        
        M_mid = M0 + Rx*y_mid - Ry*x_mid
        
        for k in range(segment_counter):
            Fx = Fx_ls[k]
            Fy = Fy_ls[k]
            xF = xF_ls[k]
            yF = yF_ls[k]
            M_mid += Fx*(y_mid-yF) - Fy*(x_mid-xF)
            
        theta_ls[i+1] = theta_ls[i] - ds*M_mid
        
        x_ls[i+1] = x_ls[i] + ds*np.cos(theta_mid)
        y_ls[i+1] = y_ls[i] + ds*np.sin(theta_mid)
        M_ls[i+1] = M_mid
        
    return x_ls, y_ls, thetaF_ls, M_ls, xF_ls, yF_ls, Fx_ls, Fy_ls


def error(param, rhoF, ell_needle, sF, L_contour):
    x_ls, y_ls, thetaF_ls, M_ls, xF_ls, yF_ls, Fx_ls, Fy_ls = \
        deflection_exact_distributed_F(param, rhoF, ell_needle, sF, L_contour)
    return y_ls[-1]**2 + param[1]**2 * (param[1]<0) + \
           thetaF_ls[0]**2 * (thetaF_ls[0]<0)
```

```python
L_contour = 10 # Contour length
sF = 0.7*L_contour # Middle position of the distributed force
                    # force application along the k-fiber contour
ell_needle = 0.1 # Arc length over which force is applied
F = 0.2 # External force
```

```python
param_init = [-0.08249719,  0.8906105]

ell_needle_set = np.linspace(0.5, 1.5, 11)

x_dat_set = []
y_dat_set = []
xF_set = []
yF_set = []
Fx_set = []
Fy_set = []

for ell_needle in ell_needle_set:

    rhoF = F/ell_needle # External force density
    args = (rhoF, ell_needle, sF, L_contour)
    out = scipy.optimize.minimize(error, param_init, args=args, method='Nelder-Mead')
    p_opt = out.x
    

    x_ls, y_ls, thetaF_ls, M_ls, xF_ls, yF_ls, Fx_ls, Fy_ls = \
        deflection_exact_distributed_F(p_opt, rhoF, ell_needle, sF, L_contour)
    
    x_dat_set.append(x_ls)
    y_dat_set.append(y_ls)
    xF_set.append(xF_ls)
    yF_set.append(yF_ls)
    Fx_set.append(Fx_ls)
    Fy_set.append(Fy_ls)
```

```python
@numba.njit
def deflection_exact(param, x_dat, y_dat, length_conserved=True, n_div=100):
    Rx, M0, Ry, theta0, F, xF = param
    
    F = np.abs(F)
    
    s_dat_ls, L_contour = contour_fn(x_dat, y_dat)    
    ds = L_contour/n_div
    if length_conserved:
        n_steps = n_div+2
    else:
        n_steps = int(1.15*n_div)
        
    
    x_ls = np.zeros(n_steps)
    y_ls = np.zeros(n_steps)
    M_ls = np.zeros(n_steps)
    theta_ls = np.zeros(n_steps)
    
    
    theta_ls[0] = theta0
    found_F = False
    yF = 0
    thetaF = np.pi/2
    
    for i in range(n_steps-1):
 
        xF_sign = np.sign(xF) * np.sign(xF - x_ls[i])
        if xF_sign < 0 and found_F == False:
            found_F = True
            
            iF = i-1
            yF = interpolate_two_pt(xF, x_ls[iF], x_ls[iF+1], y_ls[iF], y_ls[iF+1])
            thetaF = np.pi/2 + np.arctan2(y_ls[iF+1]-y_ls[iF], x_ls[iF+1]-x_ls[iF])
            Fx = F * np.cos(thetaF)
            Fy = F * np.sin(thetaF)
            
            
        M = M0 + Rx*y_ls[i] - Ry*x_ls[i]
        if found_F:
            M += Fx*(y_ls[i]-yF) - Fy*(x_ls[i]-xF)
        
        
        theta_mid = theta_ls[i] - ds/2*M
        y_mid = y_ls[i] + ds/2*np.sin(theta_mid)
        x_mid = x_ls[i] + ds/2*np.cos(theta_mid)
        
        
        M_mid = M0 + Rx*y_mid - Ry*x_mid
        if found_F:
            M_mid += Fx*(y_mid-yF) - Fy*(x_mid-xF)
            
        theta_ls[i+1] = theta_ls[i] - ds*M_mid
        x_ls[i+1] = x_ls[i] + ds*np.cos(theta_mid)
        y_ls[i+1] = y_ls[i] + ds*np.sin(theta_mid)
        M_ls[i+1] = M_mid 
        
        
    return x_ls, y_ls, M_ls, yF, thetaF
```

```python
def res_fn_exact(param, x_dat, y_dat, length_conserved=True, end_to_end=np.inf, n_dense=500):
    
    output_deflection = deflection_exact(param, x_dat, y_dat, length_conserved=length_conserved)

    x_mod = output_deflection[0]
    y_mod = output_deflection[1]

    s_mod, _ = contour_fn(x_mod, y_mod)
    
    _, L_contour = contour_fn(x_dat, y_dat)
    x_mod_truncated = x_mod[s_mod <= 0.8*L_contour]
    y_mod_truncated = y_mod[s_mod <= 0.8*L_contour]
        
    loop_length_net = loop_length_net_fn(x_mod_truncated, y_mod_truncated)

    penalty_loop = 1 + (10*loop_length_net/L_contour)**2
    
    res = distances_pair_min_new(x_dat, y_dat, x_mod, y_mod)
    
    penalty_end_match = 1+(res[-1]/end_to_end)**2
    
    res = res * penalty_end_match * penalty_loop
        
    return res
```

```python
@numba.njit
def loop_length_net_fn(x_mod, y_mod):
    ds = np.sqrt((x_mod[1]-x_mod[0])**2 + (y_mod[1]-y_mod[0])**2)
    x_mod_mat_1, x_mod_mat_2 = mesh_jit(x_mod, x_mod)
    y_mod_mat_1, y_mod_mat_2 = mesh_jit(y_mod, y_mod)

    ii, jj = mesh_jit(np.arange(len(x_mod)), np.arange(len(x_mod)))
    
    d_pairwise_mat = np.sqrt((x_mod_mat_1-x_mod_mat_2)**2 + (y_mod_mat_1-y_mod_mat_2)**2)
    
    width = np.array([0.04*len(x_mod), 2.2]).max()
    coord_diff = np.abs(ii-jj)
    ind_far = np.where((coord_diff.ravel() > width) & (coord_diff.ravel() < len(x_mod)-width))
    
    
    jj, ii = mesh_jit(np.arange(len(x_mod)), np.arange(len(x_mod)))
    jj_far = jj.ravel()[ind_far]
    ii_far = ii.ravel()[ind_far]

    dd = d_pairwise_mat.ravel()[ind_far]
    
    jj_loop = jj_far[dd < 2*ds]
    ii_loop = ii_far[dd < 2*ds]
    
    ii_groups, jj_groups = find_clusters(ii_loop, jj_loop)
    
    
    loop_net = 0
    for k in range(len(ii_groups)):
        i_low = ii_groups[k]
        i_high = jj_groups[k]
        loop_net += (i_high-i_low)*ds
        
    return loop_net


@numba.njit
def find_clusters(pos_arr, pos_arr_2):
    ii_groups = [np.int(x) for x in range(0)]
    jj_groups = [np.int(x) for x in range(0)]
    
    for i in range(len(pos_arr)):
        if pos_arr[i] > 0:
            ind = np.where( (pos_arr.ravel() <= pos_arr[i]+4) & (pos_arr.ravel() > 0))
            
            ii_groups.append(int(np.median(pos_arr[ind])))
            jj_groups.append(int(np.median(pos_arr_2[ind])))
            
            pos_arr[np.where(pos_arr <= pos_arr[i]+4)] = 0
    ii_groups = np.array(ii_groups)
    jj_groups = np.array(jj_groups)
    
    # Remove duplicates
    ind_keep = np.where(ii_groups.ravel() < jj_groups.ravel())
    ii_groups = ii_groups[ind_keep]
    jj_groups = jj_groups[ind_keep]
    
    return ii_groups, jj_groups
```

```python
p_opt_set = []
xF_inferred_set = []
F_inferred_set = []

x_mod_trunc_set = []
y_mod_trunc_set = []
thetaF_set = []
yF_inferred_set = []

RMSE_set = []

for i in tqdm(range(len(ell_needle_set))):
    x_dat = x_dat_set[i]
    y_dat = y_dat_set[i]

    Rx_init = 0
    M0_init = 0
    theta0_init = np.arctan(y_dat[y_dat==y_dat.max()]/x_dat[y_dat==y_dat.max()])[0]
    xF = x_dat[np.where(y_dat == np.max(y_dat))]
    length_conserved = False
    end_to_end = 1e-3
    
    args = (x_dat, y_dat, length_conserved, end_to_end)

    error_min = np.inf

    for F_init in np.logspace(-1, 0.5, 6):
        Ry_init = -0.1*F_init
        param_init = np.array([Rx_init, M0_init, Ry_init, theta0_init, F_init, xF])

        output = scipy.optimize.least_squares(res_fn_exact, param_init, args=args, method="lm")

        if error_min > output.cost:
            p_opt = output.x
            error_min = output.cost
    
    output_deflection = deflection_exact(p_opt, x_dat, y_dat, length_conserved)
    x_mod, y_mod, M_mod, yF, thetaF = output_deflection

    _, ind_pairs = distances_pair_min(x_dat, y_dat, x_mod, y_mod)
    ind_pairs = ind_pairs.astype('int')
    x_mod_truncated = x_mod[:ind_pairs[-1]+1]
    y_mod_truncated = y_mod[:ind_pairs[-1]+1]
    
    
    x_mod_trunc_set.append(x_mod_truncated)
    y_mod_trunc_set.append(y_mod_truncated)
    yF_inferred_set.append(yF)
    thetaF_set.append(thetaF)
    
    res_exact, _ = distances_pair_min(x_dat, y_dat, x_mod_truncated, y_mod_truncated)
    err_exact = np.sqrt(np.mean(res_exact**2))
    RMSE_set.append(err_exact)
    
    p_opt_set.append(p_opt)
    Rx, M0, Ry, theta0, F, xF = p_opt
    xF_inferred_set.append(xF)
    F_inferred_set.append(F)
```
