```python
def alpha_to_rgb(color, alpha):
    """
    Calculates the RGB code of a color with given transparency.
    
    Inputs
    ------
    color : Specified color in HEX representation.
    alpha : Number in the [0,1] interval.
    
    Outputs
    -------
    rgb_new : RGB number triplet of the color with transparency `alpha`.
    """
    rgba = np.array(colors.to_rgba(color))
    rgb_new = alpha*(rgba[0:3]-1)+1
    return rgb_new
```

### Fit errors

```python
RMSE_Mkzero_ls = df["RMSE (mcmc)"].values
RMSE_Rxzero_ls = df_Rxzero["RMSE (mcmc)"].values
RMSE_Ryzero_ls = df_Ryzero["RMSE (mcmc)"].values
RMSE_Mpzero_Mknonzero_ls = df_Mpzero_Mknonzero["RMSE (mcmc)"].values
RMSE_Rxzero_Mknonzero_ls = df_Rxzero_Mknonzero["RMSE (mcmc)"].values

plt.figure(figsize=(5,4))
seed = 4
np.random.seed(seed)

var1 = RMSE_Ryzero_ls #Rx only
var2 = RMSE_Rxzero_ls #Ry/Mp only
var3 = RMSE_Rxzero_Mknonzero_ls # Ry/Mk only
var4 = RMSE_Mpzero_Mknonzero_ls # Rx,Ry,Mk
var5 = RMSE_Mkzero_ls #Rx,Ry,Mp

plt.yscale("log")

hex_colors = ["#D8BFD8","#FF7256","#FFD700","#3CB371","#3D59AB"]
alphas = [0.6,0.6,0.6,1,1]
rgb_colors = []
for i in range(len(hex_colors)):
    rgb_colors.append(alpha_to_rgb(hex_colors[i],alphas[i]))

sns.swarmplot(data=[var1,var2,var3,var4,var5], palette = rgb_colors, size = 5)
plt.errorbar([0,1,2,3,4],[np.mean(var1),np.mean(var2),np.mean(var3),np.mean(var4),np.mean(var5)],\
             yerr = [np.std(var1)/np.sqrt(len(var1)),np.std(var2)/np.sqrt(len(var2)),\
                     np.std(var3)/np.sqrt(len(var3)),np.std(var4)/np.sqrt(len(var4)), \
                     np.std(var5)/np.sqrt(len(var5))],\
             elinewidth = 3, fmt = '_', color = 'k', marker = '.', markersize = 10, \
             capsize = 5, capthick = 1, zorder = 10)

plt.xticks([0,1,2,3,4], ("","","",""),fontsize=15)
plt.yticks(fontsize=15)

plt.tight_layout()
plt.ylim(0.0004,1)
# plt.gca().invert_yaxis()

if save_images:
    plt.savefig('fig_outputs/RMSEall.pdf')
plt.show()

import scipy.stats
from scipy.stats import mannwhitneyu
print("mann-whitney U test")
print("Mp zero",mannwhitneyu(RMSE_Mkzero_ls, RMSE_Mpzero_Mknonzero_ls, alternative='two-sided')[1])
print("Ry/Mk only",mannwhitneyu(RMSE_Mkzero_ls, RMSE_Rxzero_Mknonzero_ls, alternative='two-sided')[1])
print("Rx only",mannwhitneyu(RMSE_Mkzero_ls, RMSE_Ryzero_ls, alternative='two-sided')[1])
print("Ry/Mp only",mannwhitneyu(RMSE_Mkzero_ls, RMSE_Rxzero_ls, alternative='two-sided')[1])
```

```python
x_peak_arr = np.zeros(len(df))
for index, row in df.iterrows():
    x_dat = row["x_dat"]
    y_dat = row["y_dat"]
    x_peak_arr[index] = x_dat[np.where(y_dat==y_dat.max())][0]
    
Rx_arr = df["Rx (mcmc)"].values
M0_arr = df["M0 (mcmc)"].values
Ry_arr = df["Ry (mcmc)"].values
PP_K_angle_arr = df["PP-K angle (deg)"].values
L_arr = df["L"].values
x_peak_norm_arr = x_peak_arr/L_arr
```

### $M_p$ vs. $\hat{x}_{\rm peak}$

```python
fig = plt.figure(figsize=(6, 4.5))
plt.scatter(x_peak_norm_arr, M0_arr, 50, color='#3D59AB')

plt.axhline(0, ls='--', color='k', lw=1.00, alpha=0.3)
plt.axvline(0.499, ls='--', color='k', lw=1.00, alpha=0.3)

x_peak_low = np.array([.35,.39, .43, .47, .51])
x_peak_upp = np.array([.39,.43, .47, .51, .56])
x_peak_avg = 0.5*(x_peak_low + x_peak_upp)
n_bins = len(x_peak_low)

M0_means = np.zeros(n_bins)
M0_sem = np.zeros(n_bins)
x_peak_sem = np.zeros(n_bins)

for i in range(n_bins):
    M0_subset = M0_arr[(x_peak_norm_arr >= x_peak_low[i]) & (x_peak_norm_arr < x_peak_upp[i])]
    M0_means[i] = np.mean(M0_subset)
    M0_sem[i] = np.std(M0_subset)/np.sqrt(len(M0_subset))

    x_peak_subset = x_peak_norm_arr[(x_peak_norm_arr >= x_peak_low[i]) & (x_peak_norm_arr < x_peak_upp[i])]
    x_peak_sem[i] = np.std(x_peak_subset)/np.sqrt(len(x_peak_subset))
    
for i in range(n_bins):
    plt.errorbar(x_peak_avg[i], M0_means[i], xerr = x_peak_sem[i], yerr = M0_sem[i], elinewidth = 3, fmt = '_', \
                 color = 'k', marker = '.', markersize = 10, capsize = 5, capthick = 1, zorder = 10)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

if save_images:
    plt.savefig('fig_outputs/M0_xpeak.pdf')
```
