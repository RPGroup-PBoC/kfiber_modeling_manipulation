```python
path = "../../dat/WT_unmanip_20210722/"
spindle = "20180611_cell002" # 20210526_cell004, 20210526_cell001, 20180628_cell010, 20180611_cell002

df_filtered = df.loc[(df['Spindle'] == spindle)]

for index, row in df_filtered.iterrows():
    cell = row["Spindle"]
    spindle = row["Spindle"]
    kfiber = row["Kfiber"]
    
    df_profile = pd.read_csv("%s/%s/%s_before_%s.csv"%(path,cell,spindle,kfiber))
    x_dat = np.array(df_profile["X"].values)
    y_dat = np.array(df_profile["Y"].values)
    
    image = mpimg.imread("%s/%s/%s_before.tif"%(path,cell,spindle))
    plt.imshow(image, cmap ='gray')
    plt.plot(x_dat, y_dat, color = 'orange', lw = 3)

plt.xlim(10,160)
plt.ylim(15,100)
plt.axis("off")
plt.gca().invert_yaxis()
if save_images:
    plt.savefig('fig_outputs/20180611_cell002_image.pdf')
plt.show()

for index, row in df_filtered.iterrows():
    cell = row["Spindle"]
    spindle = row["Spindle"]
    kfiber = row["Kfiber"]

    df_profile = pd.read_csv("%s/%s/%s_before_%s.csv"%(path,cell,spindle,kfiber))
    x_dat = np.array(df_profile["X"].values)
    y_dat = np.array(df_profile["Y"].values)
    
    x_mod = row["x (exact)"]
    y_mod = row["y (exact)"]

    theta0 = np.arctan2(y_dat[3]-y_dat[0], x_dat[3]-x_dat[0])
    if theta0 < 0.0:
        y_mod = -y_mod

    if '2017' in spindle:
        scope_calibration = 0.0657
    else:
        scope_calibration = 0.105
            
    x_dat = x_dat*scope_calibration
    y_dat = y_dat*scope_calibration

    angle = np.arctan((y_dat[-1] - y_dat[0]) / (x_dat[-1] - x_dat[0]))
    x_mod,y_mod = rotate(x_mod,y_mod,-angle)

    x_mod = np.array([x_dat[0] + x for x in x_mod])
    y_mod = np.array([y_dat[0] + y for y in y_mod])

    plt.imshow(image, cmap='gray', extent=(0, scope_calibration*image.shape[0], \
                                           0, scope_calibration*image.shape[1]))
    
    plt.plot(x_dat, y_dat, color = 'orange', lw = 8)
    plt.plot(x_mod, y_mod, color = '#3D59AB', lw = 3)

plt.axis("equal")
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.gca().invert_yaxis()
if save_images:
    plt.savefig('fig_outputs/20180611_cell002_modelfits.pdf')
plt.show()
```

```python
y_max_ls = []
x_peak_ls = []
theta0_ls = []
L_ls = []

for index, row in df.iterrows():
    x_dat = row["x_dat"]
    y_dat = row["y_dat"]
    
    # Pole-kinetochore (P-K) distance
    L = x_dat[-1] - x_dat[0]

    # Initial angle in the rotated profile
    theta_init = np.arctan2(y_dat[3]-y_dat[0], x_dat[3]-x_dat[0])

    x_peak_ls.append(x_dat[np.where(y_dat==y_dat.max())][0])
    y_max_ls.append(np.max(y_dat))
    theta0_ls.append(theta_init)
    L_ls.append(L)
```

```python
Rx_ls = df["Rx (mcmc)"].values
M0_ls = df["M0 (mcmc)"].values
Ry_ls = df["Ry (mcmc)"].values
theta0_ls = df["theta0 (mcmc)"].values
RMSE_exact_ls = df["RMSE (mcmc)"].values
RMSE_exact_Rxzero_ls = df_Rxzero["RMSE (mcmc)"].values
RMSE_exact_Ryzero_ls = df_Ryzero["RMSE (mcmc)"].values
PP_K_angle_ls = df["PP-K angle (deg)"]
L_ls = df["L"].values
x_peak_norm_ls = x_peak_ls/L_ls
indices_all = np.arange(len(L_ls))

Rx_ls = Rx_ls * 1000
M0_ls = M0_ls * 100

ind_good = list(np.where(x_peak_norm_ls<0.5)[0])
# ind_good.remove(1)
# ind_good.remove(11)
# ind_good.remove(22)
# ind_good.append(12)

Rx_ls = Rx_ls[ind_good]
M0_ls = M0_ls[ind_good]
Ry_ls = Ry_ls[ind_good]
theta0_ls = theta0_ls[ind_good]
RMSE_exact_ls = RMSE_exact_ls[ind_good]
RMSE_exact_Rxzero_ls = RMSE_exact_Rxzero_ls[ind_good]
RMSE_exact_Ryzero_ls = RMSE_exact_Ryzero_ls[ind_good]
L_ls = L_ls[ind_good]
x_peak_norm_ls = x_peak_norm_ls[ind_good]
PP_K_angle_ls = PP_K_angle_ls[ind_good]

ind_outliers = np.where((x_peak_norm_ls > 0.5) | (Rx_ls < -20))[0]
ind_final = indices_all[ind_good]
```

### Average profiles

```python
nbins = 20

colors=["gold", "orange", "red", "maroon", "black"]

for i in range(n_angles):
    
    df_angle = df[(df['PP-K angle (deg)'] >= angles_low[i]) & (df['PP-K angle (deg)'] <= angles_high[i])]
    
    rot_angle = np.mean([angles_low[i], angles_high[i]])
    
    x_mod_ls, y_mod_ls = [], []
    for n, row in df_angle.iterrows():
        if not (n in ind_good):
            continue
        x_mod = row["x (exact)"]
        y_mod = row["y (exact)"]
        L = row["L"]
        
        x_rot, y_rot = rotate(x_mod, y_mod, np.radians(-rot_angle))
        x_rot_scale = x_rot/L
        y_rot_scale = y_rot/L
        
        x_mod_ls.append(x_rot_scale)
        y_mod_ls.append(y_rot_scale)
        
        plt.plot(x_rot_scale, y_rot_scale, color='grey', ls='--', alpha=0.5)
        
    x, y = [], []
    [x.extend(el) for el in x_mod_ls]
    [y.extend(el) for el in y_mod_ls]
    x = np.array(x)
    y = np.array(y)
    
    n, _ = np.histogram(x, bins=nbins)
    
    sy, _ = np.histogram(x, bins=nbins, weights=y)
    sy2, _ = np.histogram(x, bins=nbins, weights=y*y)
    mean = sy / n
    std = np.sqrt(sy2/n - mean*mean)
    plt.errorbar((_[1:] + _[:-1])/2, mean, yerr=std, fmt = 'o-', color='#333333') #colors[i])
    
plt.axhline(y=0, color="k", lw=0.5, alpha=0.3)
plt.axis("equal")
plt.xlabel('normalized position \nparallel to the pole-pole axis', fontsize=14)
plt.ylabel('normalized position \nnormal to the pole-pole axis', fontsize=14)
if save_images:
    plt.savefig('fig_outputs/average_profiles.pdf')
```
